```python
#import all the neccessary packages
import torch
import torch.nn as nn
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
```

```python
device=torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python

train_data=torchvision.datasets.MNIST(root='./data',train=True,
                                      transform=transforms.ToTensor(),download=True)
```

```python

test_data=torchvision.datasets.MNIST(root='./data',train=False,
                                      transform=transforms.ToTensor())
```

```python
#data loader
train_split=torch.utils.data.DataLoader(dataset=train_data,shuffle=True,batch_size=100)
test_split=torch.utils.data.DataLoader(dataset=test_data,shuffle=True,batch_size=100)
```

```python
class classifier(nn.Module):
    def __init__(self,input_size,hidden_size,num_of_classes):
        super(classifier,self).__init__()
        self.l1=nn.Linear(input_size,hidden_size)
        self.relu=nn.ReLU()
        self.l2=nn.Linear(hidden_size,10)

    def forward(self,x):
        out=self.l1(x)
        out=self.relu(out)
        out=self.l2(out)
        return out

```

```python
input_size=784
num_of_classes=10
hidden_size=400
```

```python
los=nn.CrossEntropyLoss()
model=classifier(input_size,hidden_size,num_of_classes)
opt=torch.optim.Adam(model.parameters(),lr=0.01)
```

```python
num_epochs=3
for i in range(num_epochs):
    for i,(images,lables) in enumerate(train_split):
        images=images.reshape(-1,28*28).to(device)
        lables=lables.to(device)

        out=model(images)
        loss=los(out,lables)

        opt.zero_grad()
        loss.backward()
        opt.step()
        if((i+1)%100)==0:
            print(f'erpocs{i},loss{loss.item()}')

```
