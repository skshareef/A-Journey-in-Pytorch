# Soft max fucntion




image.png

Squezes the ouput in between zero and one

```python
#implementing softmax function manullay
def softmax(x):
    return np.exp(x)/np.sum(np.exp(x),axis=0)

x=np.array([3.0,4.0,2.5,0.5])
outputs=softmax(x)
print('softmax outputs ',outputs)
```

```python
#calculating in pytorch
x=torch.tensor([3.0,4.0,2.5,0.5])
out=torch.softmax(x,dim=0)
print(out)
```

Soft-max usually combined with cross entropy.


1.   Used in classification problems where output lies between zero and one.
2.   Can aslo be used in multi class-label problems.



image.png



If the predciton is close to the actual value then the cross entropy loss would be small.









```python
 #in numpy
 def cross_e(a,p):
    loss=-np.sum(a*np.log(p))
    return loss
```

```python
y=np.array([1,0,1])
y_g=np.array([0.4,0.1,0.5])
y_b=np.array([0.3,0.3,0.4])
l1=cross_e(y,y_g)
l2=cross_e(y,y_b)
print(l1)
print(l2)
```

```python
#in pytorch

loss=nn.CrossEntropyLoss()

y=torch.tensor([0])

y_pred_good=torch.tensor([[2.0,1.0,0.1]])
y_pred_bad=torch.tensor([[0.5,2.0,0.1]])
```

```python
l1=loss(y_pred_good,y)
print(l1.item())
l2=loss(y_pred_bad,y)
print(l2.item())
```
