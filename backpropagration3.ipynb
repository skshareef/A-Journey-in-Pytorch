## Computational Graph
# For every operation pytorch will create a graph for us!



1.   Calculate local Gradients then we will use them to get the final Gradient.
2.   Why?      *to minimize the loss*.
3.   Forward pass-> compute Local Gradients -> Backward pass.
4.   In the backward pass compute dloss/dweights using the chain rule.







image.png

```python
 import torch as t
 # we are doing a linear regression here
 x=t.tensor(1.0)
 y=t.tensor(2.0)
 #weights
 w=t.tensor(1.0, requires_grad=True)
 #predicted
 y_hat=w*x
 loss=(y_hat-y)**2
 print(loss)

 #backward pass
 loss.backward()
 print(w.grad)

```
